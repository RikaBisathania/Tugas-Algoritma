## 5.21.4 Mengubah Data pada Pandas

```python
data_csv['Total Sepal'] = data_csv['sepal width (cm)']+data_csv['sepal length (cm)']
print(data_csv)

#Program ini secara keseluruhan adalah untuk menjumlahkan data yang ada pada 'sepal width (cm)' dan 'sepal length (cm)' sehingga akan menghasilkan data baru yang kita beri nama 'Total Sepal'
```

```python
data_csv=data_csv.drop(columns=['Total Sepal'])
print(data_csv)

#Fungsi 'drop' pada program ini adalah untuk menghapus kolom 'Total Sepal' yang sudah kita buat sebelumnya 
```

```python
data_csv=data_csv.drop(columns=['double sepal length'])
print(data_csv)

#'Drop' juga bisa menghapus kolom bawaan pada data, contohnya kolom sepal length (cm) 
```

```python
data_csv['sepal width (cm)']=data_csv['sepal width (cm)'].replace(['3.0'],'5.6')
print(data_csv)

#Perintah tersebut menggunakan metode .replace() dari pandas untuk melakukan penggantian nilai pada kolom 'sepal width (cm)'. Metode tersebut menerima dua parameter, yaitu nilai lama yang akan diganti dan nilai baru yang akan menggantikan nilai lama tersebut.
#Nilai '3.0' yang awalnya ada di kolom tersebut diganti menjadi '5.6'.
```

## 5.21.6 Menyaring Data Menggunakan Pandas 

```python
data_csv.loc[data_csv['sepal width (cm)']==3.5]

#Kode tersebut adalah sebuah perintah untuk melakukan seleksi atau filter pada dataframe 'data_csv', dengan menggunakan kondisi pada kolom 'sepal width (cm)'.
#Perintah tersebut menggunakan metode .loc() dari pandas untuk melakukan seleksi pada baris-baris yang memenuhi kondisi yang ditentukan
#Pada kasus ini, kondisi yang digunakan adalah 'sepal width (cm) == 3.5', yang artinya hanya baris-baris yang memiliki nilai kolom 'sepal width (cm)' sebesar 3.5 saja yang akan dipilih.
#Hasil dari perintah tersebut adalah dataframe yang terdiri dari baris-baris yang memenuhi kondisi tersebut, yang kemudian akan ditampilkan pada layar jika perintah tersebut diikuti dengan perintah print().
```

```python
data_csv.loc[(data_csv['sepal width (cm)']==3.5)&
             (data_csv['petal length (cm)']==1.4)]

#Perintah tersebut memilih baris-baris di mana nilai kolom 'sepal width (cm)' adalah 3.5 dan nilai kolom 'petal length (cm)' adalah 1.4. 
#Artinya, perintah ini akan mengembalikan baris-baris yang memiliki lebar kelopak bunga (sepal) sebesar 3.5 cm dan panjang kelopak bunga (petal) sebesar 1.4 cm.
```

```python
data_csv.loc[(data_csv['sepal width (cm)']==3.5)&
            (data_csv['petal length (cm)']==1.4)&
            (data_csv['petal width (cm)']>6.5)]

#Kode tersebut adalah sebuah perintah untuk melakukan seleksi atau filter pada dataframe 'data_csv', dengan menggunakan beberapa kondisi pada beberapa kolom yang berbeda.
#Perintah tersebut menggunakan metode .loc() dari pandas untuk melakukan seleksi pada baris-baris yang memenuhi semua kondisi yang ditentukan. 
#'sepal width (cm) == 3.5': Kolom 'sepal width (cm)' memiliki nilai sebesar 3.5.
#'petal length (cm) == 1.4':Kolom 'petal length (cm)' memiliki nilai sebesar 1.4.
#'petal width (cm) > 6.5': Kolom 'petal width (cm)' memiliki nilai lebih besar dari 6.5.
#Ketiga kondisi tersebut digabungkan menggunakan operator & (and) sehingga hanya baris-baris yang memenuhi semua kondisi tersebut yang akan dipilih.
```

```python
data_csv.loc[data_csv['sepal width (cm)'].str.contains(3.5)]

#Program ini hanya dapat digunakan untuk mengolah data string
#Sedangkan data yang kita miliki adalah data float
#Dengan kata lain, program ini tidak dapat digunakan untuk mengolah data_csv
```

```python
data_csv.loc[~data_csv['sepal length (cm)'].float.contains('5.1')]

#Walaupun kita sudah menggantinya menggunakan float, namun data tetap tidak bisa diolah

#Kode tersebut adalah sebuah perintah untuk melakukan seleksi atau filter pada dataframe 'data_csv', dengan menggunakan kondisi pada kolom 'sepal length (cm)' dan operator negasi (~).
#Perintah tersebut menggunakan metode .loc() dari pandas untuk melakukan seleksi pada baris-baris yang tidak memenuhi kondisi yang ditentukan.
#Pada kasus ini, kondisi yang digunakan adalah 'sepal length (cm)' tidak mengandung nilai '5.1', yang artinya hanya baris-baris yang memiliki nilai kolom 'sepal length (cm)' bukan '5.1' saja yang akan dipilih.
#Operator negasi (~) digunakan untuk membalik kondisi, sehingga hasil seleksi yang diperoleh adalah baris-baris yang tidak memenuhi kondisi tersebut.
#Namun, terdapat kesalahan pada sintaksis kode tersebut. Fungsi 'contains()' tidak tersedia untuk tipe data float pada kolom 'sepal length (cm)'. 
# Sehingga, kode tersebut tidak dapat dijalankan. 
```

```python
import re
```

```python
data_csv[data_csv["sepal length (cm)"].str.contains("3.0|3.5", regex=True)]

#Kode tersebut adalah sebuah perintah untuk melakukan seleksi atau filter pada dataframe 'data_csv', dengan menggunakan kondisi pada kolom 'sepal length (cm)' dengan menggunakan metode .str.contains() dan regular expression (regex).
#Pada kasus ini, kondisi yang digunakan adalah untuk memilih baris-baris yang mengandung nilai '3.0' atau '3.5' pada kolom 'sepal length (cm)'.
#Metode .str.contains() digunakan untuk memeriksa apakah string pada setiap elemen pada kolom 'sepal length (cm)' mengandung substring yang sesuai dengan pola yang ditentukan. 
#Pola atau pattern yang digunakan dalam kode ini adalah '3.0|3.5', yang artinya mencari substring '3.0' atau '3.5'.

#Ketika regex=True, pandas akan menafsirkan pola sebagai ekspresi reguler. 
#Sehingga, pola tersebut akan mencocokan substring
#'3.0' atau '3.5' pada setiap elemen pada kolom 'sepal length (cm)'
#Hasilnya, dataframe yang diperoleh hanya terdiri dari baris-baris yang memenuhi kondisi tersebut.
```

```python
data_csv[data_csv["sepal width (cm)"].str.contains("3.5|3.0",flags=re.I,regex=True)]

#Kode tersebut adalah sebuah perintah untuk melakukan seleksi atau filter pada dataframe 'data_csv', dengan menggunakan kondisi pada kolom 'sepal width (cm)' dengan menggunakan metode .str.contains() dan regular expression (regex).
#Pada kasus ini, kondisi yang digunakan adalah untuk memilih baris-baris yang mengandung nilai '3.0' atau '3.5' pada kolom 'sepal width (cm)'
#Metode .str.contains() digunakan untuk memeriksa apakah string pada setiap elemen pada kolom 'sepal width (cm)' mengandung substring yang sesuai dengan pola yang ditentukan.
#Pola atau pattern yang digunakan dalam kode ini adalah '3.0|3.5', yang artinya mencari substring '3.0' atau '3.5'.
#flags=re.I digunakan untuk mengabaikan case sensitivity (tidak memperhatikan huruf besar atau kecil) pada pencarian substring dengan pola yang ditentukan.
#Ketika regex=True, pandas akan menafsirkan pola sebagai ekspresi reguler. 
#Sehingga, pola tersebut akan mencocokan substring '3.0' atau '3.5' pada setiap elemen pada kolom 'sepal width (cm)'. Hasilnya, dataframe yang diperoleh hanya terdiri dari baris-baris yang memenuhi kondisi tersebut.
```

## 5.21.7 Pengelompokan Data Menggunakan Pandas

```python
data_csv.groupby(['sepal width (cm)']).mean()

#Kode tersebut adalah sebuah perintah untuk melakukan operasi grup atau pengelompokan pada dataframe 'data_csv', dengan menggunakan kolom 'sepal width (cm)' sebagai kunci pengelompokan.
#Setelah dilakukan pengelompokan, akan dilakukan perhitungan rata-rata pada setiap kelompok pada kolom-kolom numerik yang tersisa pada dataframe.
#Metode .groupby() digunakan untuk melakukan pengelompokan pada dataframe berdasarkan kolom yang ditentukan
#Pada kode ini, kolom yang digunakan sebagai kunci pengelompokan adalah 'sepal width (cm)'.
#Setelah pengelompokan dilakukan, metode .mean() dipanggil untuk menghitung nilai rata-rata pada setiap kelompok yang ada pada kolom-kolom numerik pada dataframe. 
#Hasilnya adalah sebuah dataframe baru yang berisi nilai rata-rata pada setiap kelompok pada setiap kolom numerik pada dataframe.
```

```python
data_csv.groupby(['sepal width (cm)']).mean().sort_values('petal width (cm)',ascending=True)

#Kode tersebut adalah perintah yang hampir sama dengan perintah sebelumnya, yaitu melakukan operasi grup atau pengelompokan pada dataframe 'data_csv' dengan menggunakan kolom 'sepal width (cm)' sebagai kunci pengelompokan dan melakukan perhitungan rata-rata pada setiap kelompok pada kolom-kolom numerik yang tersisa pada dataframe
#metode .sort_values() untuk mengurutkan hasil rata-rata pada kolom 'petal width (cm)' secara ascending (dari nilai terkecil ke terbesar).
#Metode .sort_values() digunakan untuk mengurutkan nilai pada dataframe berdasarkan nilai pada kolom tertentu.
#Pada kode ini, dataframe yang dihasilkan dari operasi pengelompokan diperoleh dengan memanggil metode .groupby() pada 'data_csv' dengan kolom 'sepal width (cm)' sebagai kunci pengelompokan.
#Setelah itu, hasil pengelompokan tersebut diurutkan berdasarkan nilai pada kolom 'petal width (cm)' secara ascending (dari nilai terkecil ke terbesar) dengan menggunakan metode .sort_values().
#Hasil dari perintah ini adalah sebuah dataframe baru yang berisi nilai rata-rata pada setiap kelompok pada setiap kolom numerik pada dataframe 'data_csv', diurutkan berdasarkan nilai pada kolom 'petal width (cm)' secara ascending (dari nilai terkecil ke terbesar).
```

## 5.22.4 Properti Pembuatan Histogram dengan Matplotlib

```python
Penyanyi = ['Jihyo Twice','Coldplay','Westlife','Ed Sheeran']
Penggemar = [99,49,29,150]
#Tuliskan yaitu anggota dari variabel 'Penyanyi' dan 'Penggemar'

plt.figure(figsize=(8,4))
#figsize adalah untuk menyatakan lebar, tinggi dalam inci.

histogram = plt.bar(Penyanyi,Penggemar,color='purple')
#Secara keseluruhan program ini adalah untuk menyatakan menyatakan grafik yang dibentuk adalah berwarna ungu

histogram[0].set_hatch('x')
#Perhatikan grafik yang dibentuk oleh 'Jihyo Twice' dan '100', variasinya adalah berbentuk 'x' yang dicetak berulang sehingga membentuk motif

histogram[1].set_hatch('o')
#Perhatikan grafik yang dibentuk oleh 'Coldplay' dan '60', variasinya adalah berbentuk 'o' yang dicetak berulang sehingga membentuk motif

histogram[2].set_hatch('.')
##Perhatikan grafik yang dibentuk oleh 'Westlife' dan '30', variasinya adalah berbentuk '.' yang dicetak berulang sehingga membentuk motif

histogram[3].set_hatch('/')
#Perhatikan grafik yang dibentuk oleh '' dan '60', variasinya adalah berbentuk 'o' yang dicetak berulang sehingga membentuk motif

plt.show()
#Fungsinya sama seperti print, namun disini fungsinya adalah untuk mencetak grafik
```

```python
Penyanyi = ['Jihyo Twice','Coldplay','Westlife','Ed Sheeran']
Penggemar = [99,49,29,150]
#Sama seperti pernyataan pada kolom sebelumnya, nyatakan anggota dari variabel 'Penyanyi' dan 'Penggemar'

plt.figure(figsize=(8,4))
#(8,4) : 8 adalah jumlah bagian pada garis y dan 4 adalah jumlah bagian pada garis x

hist = plt.bar(Penyanyi,Penggemar,color='orange')
#Program ini adalah untuk memuat perintah mencetak bar dengan warna orange

pola = ['x','o','.','/']
#Perhatikan anggota dari variabel penyanyi dan penggemar yang berjumlah 4, maka bar yang dibentuk dari keempat anggota variabel tersebut akan berurutan sesuai dengan perintah
#Bar yang pertama bermotif 'x', bar kedua bermotif 'o', bar ketiga bermotif '.', bar keempat bermotif '/'
for i in [0,1,2,3]:
    hist[i].set_hatch(pola[i])
#Kode tersebut mengakses objek plot histogram pada indeks ke-0, ke-1, ke-2, dan ke-3 pada list hist, kemudian mengatur pola garis pada setiap plot histogram tersebut sesuai dengan pola pada indeks ke-0, ke-1, ke-2, dan ke-3 pada list pola.
#Dalam hal ini, pola adalah sebuah list yang berisi pola garis yang akan digunakan pada plot histogram. 
#Kode hist[i].set_hatch(pola[i]) digunakan untuk mengatur pola garis pada plot histogram dengan indeks ke-i pada list hist agar sama dengan pola garis pada indeks ke-i pada list pola.
    
plt.show()
#Kode ini adalah untuk menampilkan segala sesuatu yang sudah kita setting
```
